```python
sample_size=50
sampleData = data.sample(n=sample_size, replace=False)
```

```python
sampleData["Batsman"].value_counts()
```

```python
#Gathering some important data from the required Batsman
print("The highest number of score for this batsman is: ")
print(sampleData['Runs'].max())
print("The highest Strike Rate for this batsman is: ")
print(sampleData['SR'].max())
print("The most number of 4s for this batsman is: ")
print(sampleData['4s'].max())
print("The most number of 6s score for this batsman is: ")
print(sampleData['6s'].max())
```

```python
againstindia = sampleData.query('Opposition == "v India"')
againstindia
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(21,7))
x = againstindia['Ground']
y = againstindia['Runs']
plt.xlabel('Ground Location', labelpad=20)
plt.ylabel('Runs Scored', labelpad=20)
plt.title('Runs vs The Ground', fontweight='bold', pad=25, fontsize=25)
plt.bar(x, y)
```

```python

plt.figure(figsize=(21,7))
x = sampleData['SR']
y = sampleData['Runs']
plt.xlabel('Strike Rate', labelpad=20)
plt.ylabel('Runs', labelpad=20)
plt.title('Runs vs Strike Rate', fontweight='bold', pad=25, fontsize=25)
plt.scatter(x, y, color='red', s=300, edgecolor='black')
```
